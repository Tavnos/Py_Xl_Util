## Custom reformating, renaming, counting... :

```python
# declaring variables
open_xl = px_m.open_xl()
key_xl = px_m.xl_keys()
dict_xldf = open_xl.to_dict()
ls_tot_col = [(px_m.col_id(i)) for i in key_xl]
d_dir_0, d_dir_1 =  {}, {}
# counting and renaming 
cr_rc = [{'direction': px_m.col_id('dir')[i], 
            'date_year': px_m.col_id('year')[i],
            'id_nr': px_m.col_id('id')[i], 
            'id_ammount': px_m.col_id('id').count(px_m.col_id('id')[i])
           } for i in range(len(ls_tot_col[0]))]
# formating and grouping 
for i in range(len(cr_rc)):
    if cr_rc[i]['direction'] == 0:
        d_dir_0[i] = cr_rc[i]
    if cr_rc[i]['direction'] == 1:
        d_dir_1[i] = cr_rc[i]
```

```python
# get groupped ammounts of specific values
def s_id_val(tgt_col = 0, max_val = 0):
    c_hv_e = []
    if tgt_col == 0:
        for i in d_dir_0:
            id_am_var = d_dir_0[i]['id_ammount']
            if d_dir_0[i]['date_year'] not in c_hv_e and id_am_var>=max_val:
                c_hv_e+=[ [d_dir_0[i]]]
    elif tgt_col == 1:
        for i in d_dir_1:
            id_am_var = d_dir_1[i]['id_ammount']
            if d_dir_1[i]['date_year'] not in c_hv_e and id_am_var>=max_val:
                c_hv_e+= [[d_dir_1[i]]]
    else:
        for i in d_dir_0:
            id_am_var = d_dir_0[i]['id_ammount']
            if d_dir_0[i]['date_year'] not in c_hv_e and id_am_var>=max_val:
                c_hv_e+=[ [d_dir_0[i]]]
        for i in d_dir_1:
            id_am_var = d_dir_1[i]['id_ammount']
            if d_dir_1[i]['date_year'] not in c_hv_e and id_am_var>=max_val:
                c_hv_e+= [[d_dir_1[i]]]
    return c_hv_e
def df_gs(tgt_col_val=0,tgt_max_val=0):
    c_h_v = s_id_val(tgt_col_val,tgt_max_val)
    d_h_v = [(c_h_v[i][0]) for i in range(len(c_h_v))]
    return px_m.df(d_h_v)
```

# Saving these to multiple sheets

```python
custom_format_total = px_m.df(cr_rc)
format_specific_group_a = px_m.df(d_dir_0)
format_specific_group_b = px_m.df(d_dir_1)
format_select_group = df_gs(2,5)
with pd.ExcelWriter('example_c.xlsx') as wr:
    custom_format_total.to_excel(wr, sheet_name='custom_format_total')
    format_specific_group_a.to_excel(wr, sheet_name='specific_group_a')
    format_specific_group_b.to_excel(wr, sheet_name='specific_group_b')
    format_select_group.to_excel(wr, sheet_name='select_group')
```
